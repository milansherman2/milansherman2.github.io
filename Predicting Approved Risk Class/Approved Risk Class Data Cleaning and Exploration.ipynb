```python
from sklearn.ensemble import IsolationForest
import numpy as np

num_df = df_num[['age', 'height', 'weight', 'bmi_app_state', 'income_app_state']]

X = num_df.values
# Create an instance of the IsolationForest algorithm
clf = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)

# Fit the model to the data
clf.fit(X)
y_pred = clf.predict(X)

# Print the indices of the predicted outliers
outlier_indices = np.where(y_pred == -1)[0]
outliers_df = num_df.iloc[outlier_indices]

print(outliers_df)
```

```python
# Define the relative_risk_map dictionary
relative_risk_map = {
    'super_elite_nt':0.70,
    'elite_nt':0.80,
    'preferred_nt':0.90,
    'select_nt':1.23,
    'essential_nt':1.45,
    'essential_plus_nt':2.00,
    'select_tb':2.25,
    'essential_tb':2.50,
    'essential_plus_tb':2.65,
    'ineligible':2.75, 
    'declined':3.00,
    'not_approved' : 3.00
}


# Map the 'approved risk class' column to the dictionary values and create a new column 'relative_risk'
df['risk'] = df['approved_risk_class'].map(relative_risk_map)
df.head()
```

```python
df['risk'].isna().sum()
```

```python
risk = df.drop(columns = ['approved_risk_class', 'mortality_rate'], axis=1)
risk.head()
```

```python
from category_encoders import TargetEncoder
import pandas as pd


# Define the features and target variables
X = risk.drop(columns=['risk'])
y = risk['risk']

# Fit and transform the categorical features using TargetEncoder
enc = TargetEncoder(min_samples_leaf=20, smoothing=10).fit(X, y)
X_transformed = enc.transform(X)

# Concatenate the transformed categorical features with the original numerical target variable
risk_transformed = pd.concat([X_transformed, y], axis=1)

# Print the resulting dataframe
# print(risk_transformed.head())

risk_transformed.to_csv('risk_regression.csv')
```
